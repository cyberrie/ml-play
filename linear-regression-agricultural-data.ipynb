```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
# read the data
data=pd.read_csv("agricultural_data.csv")
print('CSV loaded')
```

```python
# group the data by Crop Type and calculate the mean Crop Yield for each type
grouped_data=data.groupby("Crop Type")['Crop Yield (tons/ha)'].mean()
```

```python
# bar chart showing the avg crop yield for each crop type
plt.bar(data['Crop Type'], data['Crop Yield (tons/ha)'], color='maroon')

# add title and labels
plt.title('Total Sales by Product Category')
plt.xlabel('Crop Type')
plt.ylabel('Crop Yield (tons/ha)')

# show the plot
plt.show()
```

```python
# scatter plot showing relationship between Fertiliser User (kg/ha) and Crop Yield (tons/ha)
plt.scatter(data['Fertiliser Used (kg/ha)'], data['Crop Yield (tons/ha)'], color='steelblue')

plt.title('Relationship between Fertiliser User (kg/ha) and Crop Yield (tons/ha)')
plt.xlabel('Fertiliser Used (kg/ha)')
plt.ylabel('Crop Yield (tons/ha)')

plt.show()
```

```python
# histogram showing distribution of Sunlight Hours across all fields
plt.hist(data['Sunlight Hours'], color='gold')

plt.title('Distribution of Sunlight Hours across all fields')
plt.xlabel('Sunlight Hours')
plt.ylabel('Frequency')

plt.show()
```
