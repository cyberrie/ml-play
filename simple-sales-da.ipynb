```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read the data
data=pd.read_csv("retail_data.csv")
print('CSV file loaded')
```

```python
# group the data by prouct category and sum the total sales
sales_by_cat=data.groupby('Product Category')['Total Sales'].sum()
```

```python
# bar chart showing total sales for each product category across all stores
plt.bar(data['Product Category'], data['Total Sales'])

# add title and labels
plt.title('Total Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')

# show the plot
plt.show()
```

```python
# scatter plot Advertising Spend vs. Total Sales
plt.scatter(data['Advertising Spend'], data['Total Sales'])

# title and labels
plt.title("Relationship between Advertising Spend and Total Sales")
plt.xlabel('Advertising Spend')
plt.ylabel('Total Sales')

plt.show()
```

```python
# histogram to visualise distribution of customers across all stores
plt.hist(data['Number of Customers'])

plt.title("Customer Distribution across All Stores")
plt.xlabel('Number of Customers')
plt.ylabel('Frequency')

plt.show()
```
