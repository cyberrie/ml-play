```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
# read the data
data=pd.read_csv("agricultural_data.csv")
print('CSV loaded')
```

```python
# linear regression model predicting Crop Yield based on Water Supplied (litres/ha)
X=data[['Water Supplied (litres/ha)']]
y=data['Crop Yield (tons/ha)']

# create LG model
model=LinearRegression()
model.fit(X,y)

# predict Y
y_pred=model.predict(X)

# plot the data
plt.scatter(X, y, color='navy') #data points
plt.plot(X, y_pred, color='lightcoral') # regression line
plt.title('Crop Yield Prediction based on Water Supplied (litres/ha)')
plt.xlabel('Water Supplied (litres/ha)')
plt.ylabel('Crop Yield (tons/ha)')

# calculate R
r_squared=r2_score(y, y_pred)
print(f'R-squared = {r_squared:.4f}')

plt.show()
```

```python
# no correlation based on R so not really worth predicting?? unsure..
# predict crop yield based on new water supplied volume
new_water_supplied=np.array([[20000]])
predicted_crop_yield=model.predict(new_water_supplied)

print(f'Predicted Crop Yield for {new_water_supplied[0][0]} litres/ha of water supplied: {predicted_crop_yield[0]:.2f} tons/ha')
```
